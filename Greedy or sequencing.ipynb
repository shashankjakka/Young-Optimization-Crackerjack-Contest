```python
## Greedily choose max run products initially and then onto products with minimum changeover time and in a tie 
## situation product with maximum run time is choosen again

def greedy(line):
    df=d[['Pid',line]]
    df=df[df[line]!=0]
    df.reset_index(drop=True,inplace=True)
    df=df.sort_values(line,ascending=False)
    df.reset_index(drop=True,inplace=True)
    sequence=[df.iloc[0,0]]
    Days=[]
    for i in range(0,len(df)-1):
        p=sequence[-1]
        df=df[df.Pid!=p]
        temp=chdays[(chdays.From==p) & (chdays.To!=p)]

        ind=[]
        for i in list(temp.To):
            if i in list(df.Pid):
                ind.append(True)
            else: ind.append(False)
        temp=temp[ind]
        temp=temp.rename(columns={'To':'Pid'})
        temp=temp.merge(df,on='Pid')
        want=temp.iloc[np.argmax(temp[temp.Days==temp['Days'].min()][line]),1]
        Days.append(temp[temp.Days==temp['Days'].min()]['Days'].unique()[0])
        print want
        sequence.append(want)

    print sequence
    print Days
    print 20+np.sum(Days)
```
